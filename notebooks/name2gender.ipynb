**Попробуем предсказать пол по комментарию**

```python
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/boston-airbnb-reviews-a-comprehensive-overview/reviews.csv')
```

```python
df
```

**Нас интересуют колонки "reviewer_name" и "comments"**

```python
df = df[['reviewer_name', 'comments']]
df = df[df['reviewer_name'].notna()]
df.dropna(inplace=True)
```

```python
df
```

```python
df.reviewer_name.value_counts().head(40)
```

**Отберем по 10 популярных мужских и женских имен, и трансформируем их в признак мужчина/женщина**

```python
male = ['Michael', 'David', 'John', 'Chris', 'Daniel', 'Andrew', 'Mark', 'James', 'Matthew', 'Alex']
female = ['Sarah', 'Jennifer', 'Jessica', 'Laura', 'Emily', 'Lisa', 'Rachel', 'Amanda', 'Elizabeth', 'Katie']
names = male + female
```

```python
df['gender'] = df.reviewer_name.apply(lambda x: 'male' if x in male else 'female')
df = df[df.reviewer_name.isin(names)]
df = df[['comments', 'gender']]
dfmale = df.loc[df.gender == 'male']
dffemale = df.loc[df.gender == 'female']
dfmale = dfmale.sample(8440) # Сбалансируем данные
dfx = pd.concat([dfmale, dffemale])
```

```python
dfx
```

```python
dfx.to_csv('genderbased_v2.csv')
```
